# Exploratory Data Analysis of the 1Y OTPW Data

## Summary description of each table in English.
- Airline On-Time Performance Data (Flights): The Airline On-Time Performance Data is provided by the Bureau of Transportation Statistics (BTS). It contains historical flight performance records for U.S. domestic flights. Each row represents an individual flight, including scheduled and actual departure/arrival times, delays and their causes (such as carrier, weather, or NAS), flight details (airline, origin, destination), and cancellation information. This dataset is important in analyzing flight delays and more importantly in understanding patterns across airlines, routes, and times of day.

- Weather Data: The Weather Data is provided from NOAA. It consists of historical weather observations across the U.S.A, including temperature, humidity, precipitation, wind speed, visibility, and weather events such as storms and fog. The data is available on an hourly, daily or monthly basis per station, making it a key input for predicting delays influenced by weather conditions.

- Weather Stations Data: It provides metadata about weather stations, including station IDs, latitude, longitude, elevation, and names, enabling a link between weather observations and specific geographic locations. It is useful in associating weather data with airports.

- Airport Codes Data: Sourced from Airport Codes Dataset (DataHub), it contains details for airports worldwide, such as IATA and ICAO codes, airport names, cities, countries, and geographic coordinates. This dataset allows mapping airport codes in flight data to their geographic positions, supporting geospatial analysis and merging with weather data.

```python
import pyspark.sql.functions as F
import pyspark.sql.window as W
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
import pandas as pd
import sys

df = spark.read.parquet(f"{folder_path}/otpw_12m.parquet")

df = df.withColumn("label", when(col("DEP_DELAY").cast("double") > 15, 1).otherwise(0))

carrier_delay = df.groupBy("OP_UNIQUE_CARRIER").agg(F.mean("label").alias("CARRIER_DELAY_RATE"))
origin_delay = df.groupBy("ORIGIN").agg(F.mean("label").alias("ORIGIN_DELAY_RATE"))
dest_delay = df.groupBy("DEST").agg(F.mean("label").alias("DEST_DELAY_RATE"))
df = df.join(carrier_delay, on="OP_UNIQUE_CARRIER", how="left")
df = df.join(origin_delay, on="ORIGIN", how="left")
df = df.join(dest_delay, on="DEST", how="left")

df = df.withColumn("DEP_HOUR", (col("CRS_DEP_TIME").cast("int") / 100).cast("int"))
df = df.withColumn("IS_WEEKEND", when(col("DAY_OF_WEEK").isin("6", "7"), 1).otherwise(0))
df = df.withColumn("IS_WINTER", when(col("MONTH").isin("12", "1", "2"), 1).otherwise(0))
df = df.withColumn("IS_SPRING", when(col("MONTH").isin("3", "4", "5"), 1).otherwise(0))
df = df.withColumn("IS_SUMMER", when(col("MONTH").isin("6", "7", "8"), 1).otherwise(0))
df = df.withColumn("IS_FALL", when(col("MONTH").isin("9", "10", "11"), 1).otherwise(0))

holiday_dates = [
    "2015-01-01",  # New Year's Day
    "2015-01-19",  # Martin Luther King Jr. Day
    "2015-02-14",  # Valentine's Day
    "2015-02-16",  # Presidents Day
    "2015-05-25",  # Memorial Day
    "2015-07-04",  # Independence Day
    "2015-09-07",  # Labor Day
    "2015-11-26",  # Thanksgiving
    "2015-12-25",  # Christmas
    "2015-12-31"   # New Year's Eve
]
df = df.withColumn("IS_HOLIDAY", when(col("FL_DATE").isin(holiday_dates), 1).otherwise(0))

w_tailnum = W.Window.partitionBy("TAIL_NUM").orderBy("FL_DATE", "CRS_DEP_TIME")
w_airport_history = W.Window.partitionBy("ORIGIN").orderBy("FL_DATE", "CRS_DEP_TIME").rowsBetween(-sys.maxsize, -1)
w_dest_history = W.Window.partitionBy("DEST").orderBy("FL_DATE", "CRS_DEP_TIME").rowsBetween(-sys.maxsize, -1)

df = df.withColumn("Prev_TaxiIn", F.lag("TAXI_IN").over(w_tailnum))
df = df.withColumn("Prev_TaxiOut", F.lag("TAXI_OUT").over(w_tailnum))
df = df.withColumn("Prev_ArrDelay", F.lag("ARR_DELAY").over(w_tailnum))
df = df.withColumn("Prev_ArrTime", F.lag("ARR_TIME").over(w_tailnum))
df = df.withColumn("Turnaround_Time", (col("CRS_DEP_TIME").cast("int") - col("Prev_ArrTime").cast("int")))

df = df.withColumn("DEP_DATETIME", F.concat_ws(" ", col("FL_DATE"), F.format_string("%04d", col("CRS_DEP_TIME").cast("int"))))
df = df.withColumn("DEP_TIMESTAMP", F.unix_timestamp("DEP_DATETIME", "yyyy-MM-dd HHmm"))

w_airport_2h = (
    W.Window.partitionBy("ORIGIN")
    .orderBy("DEP_TIMESTAMP")
    .rangeBetween(-7200, -1)
)
df = df.withColumn(
    "Num_airport_wide_delays",
    F.sum(F.when(col("DEP_DELAY").cast("double") > 15, 1).otherwise(0)).over(w_airport_2h)
)
df = df.withColumn(
    "Num_airport_wide_cancelations",
    F.sum(F.when(col("CANCELLED").cast("double") == 1, 1).otherwise(0)).over(w_airport_2h)
)

df = df.withColumn(
    "Oncoming_flights",
    F.sum(
        F.when(
            (col("ARR_TIME").cast("int") >= (col("CRS_DEP_TIME").cast("int") - 200)) &
            (col("ARR_TIME").cast("int") <= col("CRS_DEP_TIME").cast("int")), 1
        ).otherwise(0)
    ).over(W.Window.partitionBy("ORIGIN", "FL_DATE"))
)
df = df.withColumn(
    "OntimeArrivalPct",
    F.avg(F.when(col("ARR_DELAY").cast("double") <= 0, 1).otherwise(0)).over(w_dest_history)
)
df = df.withColumn(
    "OntimeDeparturePct",
    F.avg(F.when(col("DEP_DELAY").cast("double") <= 0, 1).otherwise(0)).over(w_airport_history)
)

base_features = [ 
    "label",
    "YEAR", "QUARTER", "MONTH", "DAY_OF_MONTH", "DAY_OF_WEEK",
    "CRS_DEP_TIME", "CRS_ARR_TIME", "DISTANCE",
    "ORIGIN_AIRPORT_ID", "DEST_AIRPORT_ID",
    "ORIGIN_WAC", "DEST_WAC",
    "HourlyPrecipitation", "HourlyVisibility", "HourlyWindSpeed",
    "HourlyWindGustSpeed", "HourlyWindDirection",
    "HourlyDryBulbTemperature", "HourlyDewPointTemperature", "HourlyWetBulbTemperature",
    "HourlyRelativeHumidity", "HourlySeaLevelPressure",
    "HourlyPresentWeatherType"
]
engineered_features = [
    "DEP_HOUR", "IS_WEEKEND", "IS_WINTER", "IS_SPRING", "IS_SUMMER", "IS_FALL", "IS_HOLIDAY",
    "CARRIER_DELAY_RATE", "ORIGIN_DELAY_RATE", "DEST_DELAY_RATE",
    "Prev_TaxiIn", "Prev_TaxiOut", "Prev_ArrDelay", "Turnaround_Time",
    "Num_airport_wide_delays", "Num_airport_wide_cancelations", "Oncoming_flights",
    "OntimeArrivalPct", "OntimeDeparturePct"
]
all_cols = base_features + engineered_features

for col_name in all_cols:
    df = df.withColumn(col_name + "_dbl", F.col(col_name).cast("double"))

null_threshold = 1000000
null_counts = df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df.columns]).collect()[0].asDict()
sparse_cols = [col for col, count in null_counts.items() if count > null_threshold]
df = df.drop(*sparse_cols)

df_corr = df.dropna()

corr_cols = [c + "_dbl" for c in all_cols if c + "_dbl" in df_corr.columns]

# Sample for speed
corr_limit = 5000
df_corr_sample = df_corr.select(corr_cols).limit(corr_limit)

# Vectorize for Spark correlation
assembler = VectorAssembler(inputCols=corr_cols, outputCol="features_corr")
df_corr_vec = assembler.transform(df_corr_sample)

# Spark-based correlation computation
pearson_corr = Correlation.corr(df_corr_vec, "features_corr", "pearson").head()[0].toArray()

# Convert result to Pandas DataFrame for plotting
corr_df = pd.DataFrame(pearson_corr, columns=corr_cols, index=corr_cols)

# Plot
plt.figure(figsize=(24, 20))
sns.heatmap(
    corr_df,
    cmap='coolwarm',
    center=0,
    annot=True,
    fmt=".2f",
    annot_kws={"size": 6},
    linewidths=0.5,
    cbar_kws={"shrink": 0.7}
)
plt.title("Correlation Matrix Heatmap: Flight Delay and Features", fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
# Analyze Cardinality of Categorical Columns:

from pyspark.sql.types import StringType
from pyspark.sql import functions as F
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("=== Analyzing Cardinality of Categorical Columns ===")

# Define cardinality threshold
cardinality_threshold = 100

# Use your actual dataframe (choose the appropriate one based on pipeline stage)
df_to_analyze = otpw_3m_clean_no_leakage  # or otpw_3m_imputed

cardinality_records = []

# Identify all StringType (categorical) columns
categorical_cols = [field.name for field in df_to_analyze.schema.fields if isinstance(field.dataType, StringType)]
print(f"\nFound {len(categorical_cols)} categorical columns:")
print(categorical_cols)

# Compute cardinality for each categorical column
print("\nComputing cardinality...")
for col_name in categorical_cols:
    distinct_count = df_to_analyze.select(F.countDistinct(col_name).alias("distinct_count")).collect()[0]["distinct_count"]
    cardinality_records.append({
        "Column": col_name,
        "DistinctCount": distinct_count,
        "High_Cardinality": "Yes" if distinct_count > cardinality_threshold else "No"
    })
    print(f"  {col_name}: {distinct_count:,} distinct values")

# Convert results to Pandas
df_cardinality = pd.DataFrame(cardinality_records)
df_cardinality = df_cardinality.sort_values("DistinctCount", ascending=False)

# Show all categorical columns with their cardinality
print("\n=== All Categorical Columns ===")
display(df_cardinality)

# Filter for high cardinality only
high_cardinality_df = df_cardinality[df_cardinality["DistinctCount"] > cardinality_threshold]

print(f"\n=== High Cardinality Columns (>{cardinality_threshold} distinct values) ===")
if len(high_cardinality_df) > 0:
    display(high_cardinality_df)
    print(f"\n⚠️  Warning: {len(high_cardinality_df)} columns have high cardinality")
    print("   Consider these strategies:")
    print("   - Drop columns with extremely high cardinality (e.g., TAIL_NUM)")
    print("   - Use target encoding or frequency encoding instead of one-hot encoding")
    print("   - Group rare categories into 'Other' category")
else:
    print("✓ No high cardinality columns found")

# Plot cardinality values
plt.figure(figsize=(14, 6))
colors = ['red' if x > cardinality_threshold else 'steelblue' for x in df_cardinality['DistinctCount']]
plt.bar(range(len(df_cardinality)), df_cardinality['DistinctCount'], color=colors)
plt.axhline(cardinality_threshold, color='darkred', linestyle='--', linewidth=2, 
            label=f'High Cardinality Threshold = {cardinality_threshold}')
plt.xticks(range(len(df_cardinality)), df_cardinality['Column'], rotation=90)
plt.xlabel("Categorical Columns")
plt.ylabel("Number of Distinct Values (log scale)")
plt.yscale('log')
plt.title("Cardinality of Categorical Columns")
plt.legend()
plt.tight_layout()
plt.show()

# Recommendations based on cardinality
print("\n=== Recommendations for Feature Engineering ===")
low_card = df_cardinality[df_cardinality["DistinctCount"] <= 20]
medium_card = df_cardinality[(df_cardinality["DistinctCount"] > 20) & (df_cardinality["DistinctCount"] <= cardinality_threshold)]
high_card = df_cardinality[df_cardinality["DistinctCount"] > cardinality_threshold]

if len(low_card) > 0:
    print(f"\n✓ Low Cardinality (≤20): {len(low_card)} columns")
    print(f"  Columns: {', '.join(low_card['Column'].tolist())}")
    print("  → Use One-Hot Encoding")

if len(medium_card) > 0:
    print(f"\n⚠️  Medium Cardinality (21-{cardinality_threshold}): {len(medium_card)} columns")
    print(f"  Columns: {', '.join(medium_card['Column'].tolist())}")
    print("  → Use One-Hot Encoding (but watch for dimensionality)")

if len(high_card) > 0:
    print(f"\n❌ High Cardinality (>{cardinality_threshold}): {len(high_card)} columns")
    print(f"  Columns: {', '.join(high_card['Column'].tolist())}")
    print("  → Consider:")
    print("     • Target/Frequency Encoding")
    print("     • Dropping the column")
    print("     • Grouping rare categories")
    print("     • Feature hashing")
```

```python
from pyspark.sql import functions as F
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Group by carrier and compute stats
avg_distance_df = df_otpw_cleaned.groupBy("OP_CARRIER").agg(
    F.avg("DISTANCE").alias("avg_distance"),
    F.stddev("DISTANCE").alias("std_distance"),
    F.count("DISTANCE").alias("n")
).orderBy("avg_distance", ascending=False)

# Step 2: Convert to pandas
avg_distance_pd = avg_distance_df.toPandas()

# Step 3: Compute 95% confidence interval
avg_distance_pd["ci95"] = 1.96 * avg_distance_pd["std_distance"] / np.sqrt(avg_distance_pd["n"])

# Step 4: Plot using same styling
plt.figure(figsize=(12, 6))

# Use default Seaborn colors with black edge
ax = sns.barplot(
    data=avg_distance_pd,
    x="OP_CARRIER",
    y="avg_distance",
    ci=None,
)

# Manually add error bars
ax.errorbar(
    x=np.arange(len(avg_distance_pd)),
    y=avg_distance_pd["avg_distance"],
    yerr=avg_distance_pd["ci95"],
    fmt='none',
    c='black',
    capsize=5
)

# Final touches
plt.title("Average Distance by Carrier (with 95% CI)")
plt.xlabel("Carrier")
plt.ylabel("Average Distance (miles)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from pyspark.sql.functions import col, mean, stddev, count
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

weather_cols = ["HourlyVisibility", "HourlyWindSpeed", "HourlyPrecipitation"]

# Units for labeling
weather_units = {
    "HourlyVisibility": "Hourly Visibility (miles)",
    "HourlyWindSpeed": "Hourly Wind Speed (mph)",
    "HourlyPrecipitation": "Hourly Precipitation (inches)"
}

# For each weather variable
for wcol in weather_cols:
    # Step 1: Bin the weather variable (optional: rounding to nearest int)
    df_weather_delay = df_otpw.withColumn(f"{wcol}_bin", col(wcol).cast("int"))
    
    # Step 2: Group by binned weather value and compute stats
    agg = df_weather_delay.groupBy(f"{wcol}_bin").agg(
        mean("DEP_DELAY").alias("AvgDepDelay"),
        stddev("DEP_DELAY").alias("StdDevDepDelay"),
        count("DEP_DELAY").alias("Count")
    ).orderBy(f"{wcol}_bin")

    # Step 3: Convert to pandas
    pdf_agg = agg.toPandas().dropna()

    # Step 4: Compute 95% confidence intervals
    z = 1.96  # 95% confidence
    pdf_agg["StdErr"] = pdf_agg["StdDevDepDelay"] / np.sqrt(pdf_agg["Count"])
    pdf_agg["CI95"] = z * pdf_agg["StdErr"]

    # Step 5: Plot with confidence intervals
    plt.figure(figsize=(10, 6))
    plt.errorbar(
        pdf_agg[f"{wcol}_bin"],
        pdf_agg["AvgDepDelay"],
        yerr=pdf_agg["CI95"],
        fmt='-o',
        capsize=4,
        label="Avg Departure Delay"
    )
    plt.title(f"Average Departure Delay by {wcol} (95% CI)")
    plt.xlabel(weather_units.get(wcol, wcol))  # Use label with units if available
    plt.ylabel("Avg Departure Delay (minutes)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_list = [("df_otpw", df_otpw)]
for name, df in df_list:
    print(f"\n📊 Processing dataset: {name}")

    # Step 1: Define numeric columns using .dtypes
    numeric_cols = [col for col, dtype in df.dtypes if dtype in ("int", "bigint", "double")]

    print(f"✅ Number of numeric columns: {len(numeric_cols)}")
    print("🧪 Sample numeric columns:", numeric_cols[:10])  # optional preview

    if len(numeric_cols) == 0:
        print("⚠️ No numeric columns found. Skipping correlation matrix.")
        continue

    # Step 2: Assemble numeric columns into feature vector
    assembler = VectorAssembler(inputCols=numeric_cols, outputCol="features")
    df_vector = assembler.transform(df.select(numeric_cols).na.drop())

    # Step 3: Compute correlation matrix
    corr_matrix_row = Correlation.corr(df_vector, "features").head()
    corr_matrix = corr_matrix_row[0]

    # Step 4: Convert to Pandas DataFrame
    corr_array = corr_matrix.toArray()
    corr_df = pd.DataFrame(corr_array, index=numeric_cols, columns=numeric_cols)

    # Step 5: Plot heatmap
    plt.figure(figsize=(15, 15))
    sns.heatmap(
        corr_df,
        annot=False,
        cmap="coolwarm",
        vmin=-1, vmax=1,
        square=True,
        linewidths=0.5,
        cbar_kws={"shrink": 0.8}
    )
    plt.title(f"🔗 Correlation Matrix Heatmap: {name}", fontsize=16)
    plt.xticks(rotation=90, fontsize=8)
    plt.yticks(rotation=0, fontsize=8)
    plt.tight_layout()
    plt.show()
```

```python
import pyspark.sql.functions as F
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_final = df_5y

# Derive dep_hour_int from DEP_TIME if not already present (assumes DEP_TIME is in HHMM format)
df_final = df_final.filter(F.col("DEP_TIME").isNotNull())
df_final = df_final.withColumn("dep_hour_int", F.floor(F.col("DEP_TIME") / 100))

# Create an indicator for flights delayed at least 15 minutes.
df_delayed = df_final.withColumn("delayed_15", F.when(F.col("DEP_DELAY") >= 15, 1).otherwise(0))

# Group by dep_hour_int and DAY_OF_WEEK and calculate the percentage of flights delayed.
df_pct = df_delayed.groupBy("dep_hour_int", "DAY_OF_WEEK") \
    .agg((F.sum("delayed_15") / F.count("*") * 100).alias("pct_delayed"))

# Convert the grouped DataFrame to a Pandas DataFrame.
pdf_pct = df_pct.toPandas()

# Pivot the DataFrame so rows are dep_hour_int and columns are DAY_OF_WEEK.
heat_pct = pdf_pct.pivot(index="dep_hour_int", columns="DAY_OF_WEEK", values="pct_delayed").fillna(0)

# Set a Seaborn theme.
sns.set_theme(style="whitegrid", context="talk", font_scale=1.0)

# Use a one-color red gradient. "Reds" is a built-in matplotlib colormap.
cmap = "Reds"

plt.figure(figsize=(12,8))
ax = sns.heatmap(heat_pct, annot=True, fmt=".0f", cmap=cmap, linewidths=0.5, cbar_kws={'format': '%.0f'})
ax.set_title("% of Flights Delayed ≥ 15 Minutes", fontsize=20)
ax.set_xlabel("Day of Week", fontsize=14)
ax.set_ylabel("Hour of Day", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Build Graph & Compute PageRank

from graphframes import GraphFrame
from pyspark.sql.functions import col

# Build Graph from train_split only
vertices = train_split.selectExpr("ORIGIN as id")\
    .union(train_split.selectExpr("DEST as id"))\
    .distinct()

edges = train_split.groupBy("ORIGIN", "DEST").count()\
    .withColumnRenamed("ORIGIN", "src")\
    .withColumnRenamed("DEST", "dst")\
    .withColumnRenamed("count", "weight")

#  Create GraphFrame and compute PageRank
airport_graph = GraphFrame(vertices, edges)
pagerank_results = airport_graph.pageRank(resetProbability=0.15, maxIter=10)

#  Extract PageRank scores
pagerank_scores = pagerank_results.vertices\
    .select("id", "pagerank")\
    .withColumnRenamed("id", "ORIGIN")
```

```python
from pyspark.sql.functions import col, lit, when

# Compute threshold
hub_threshold = pagerank_scores.approxQuantile("pagerank", [0.9], 0.01)[0]

#  Create IS_HUB column
pagerank_with_hub = pagerank_scores\
    .withColumn("IS_HUB", when(col("pagerank") >= hub_threshold, lit(1)).otherwise(lit(0)))

for df_name in ["train_split", "val_split", "internal_test", "df_test"]:
    for colname in ["pagerank", "IS_HUB"]:
        if colname in eval(df_name).columns:
            exec(f"{df_name} = {df_name}.drop('{colname}')")

#  Join pagerank AND IS_HUB back into splits
cols_to_join = ["ORIGIN", "pagerank", "IS_HUB"]
for name in ["train_split", "val_split", "internal_test", "df_test"]:
    df = eval(name).drop(*[c for c in ["pagerank", "IS_HUB"] if c in eval(name).columns])
    df = df.join(pagerank_with_hub.select(*cols_to_join), on="ORIGIN", how="left")
    exec(f"{name} = df")
```
