## 3. Project Structures

### 3.1 Project Abstract:

Flight delays impose significant operational costs and customer dissatisfaction across the aviation industry. We are a data science startup tasked by an airline consortium to help minimize overhead costs and reduce inconvenience to customers, flight crew, and ground staff by accurately predicting flight delays 2 hours before scheduled departure time. Early and accurate delay predictions enable airlines to proactively manage crew scheduling, gate assignments, and passenger notifications, ultimately reducing disruption costs and improving operational efficiency. This project aims to predict whether a domestic U.S. flight will experience a departure delay of 15 minutes or more using information available at the 2-hour cutoff.

We use the OTPW (On-Time Performance and Weather) dataset, which combines flight data from the U.S. Department of Transportation's Bureau of Transportation Statistics with weather observations from the National Oceanic and Atmospheric Administration. Our initial analysis focuses on a 3-month sample from 2015 containing over 1.4 million flights across 216 features. Exploratory data analysis reveals key patterns: delays accumulate throughout the day, increasing from approximately 12% in early morning to 27% by late afternoon; the target variable exhibits significant class imbalance with roughly 80% on-time flights; and weather features such as visibility, precipitation, and wind speed show moderate correlation with delay occurrence. We systematically removed over 100 columns with excessive missingness (>50% null values) and all post-flight features to prevent data leakage.

Our primary evaluation metric is F2-score, which prioritizes recall over precision. In the operational context of flight delays, false negatives (missing true delays) are more costly than false positives, as undetected delays lead to cascading disruptions, inefficient resource allocation, and poor passenger experiences. We plan to implement an ensemble classification approach using Random Forest and Gradient Boosted Trees (XGBoost), evaluated through time-series cross-validation to maintain temporal ordering and prevent look-ahead bias. Our feature engineering pipeline will create temporal features (hour, day-of-week, seasonal indicators, holiday windows), lag-based aircraft features tracking previous flight performance and turnaround times, rolling airport congestion metrics computed over backward-looking 24-hour windows, and composite weather severity indices. All features strictly use information available at least 2 hours before scheduled departure. After validating our approach on the 3-month sample, we will scale to the full multi-year dataset while monitoring model performance, class imbalance handling strategies, and computational efficiency on distributed computing infrastructure.

### 3.3 The Data:

**Dataset Overview:**

We utilize the OTPW (On-Time Performance and Weather) dataset, a pre-joined dataset available through the Databricks file system (dbfs:/mnt/mids-w261/) that combines domestic U.S. flight data from the Bureau of Transportation Statistics with synchronized weather observations from the National Oceanic and Atmospheric Administration. This integrated dataset provides a comprehensive view of flight operations and environmental conditions, enabling us to build predictive models that account for both operational and meteorological factors.

Our initial analysis focuses on a 3-month sample from 2015 (January-March) containing **1,401,363 flights** across **216 features**. This sample allows us to develop and validate our pipeline efficiently before scaling to the full multi-year dataset. The data is stored at: `dbfs:/mnt/mids-w261/OTPW_3M_2015.csv` for the 3-month sample, with larger samples (6-month, 1-year, and full 5-year datasets) available for subsequent phases at `dbfs:/mnt/mids-w261/OTPW_60M/`.

**Data Sources:**

The OTPW dataset integrates three primary data sources:

1. **Airlines Data**: Flight operational records from the Bureau of Transportation Statistics including scheduled and actual departure/arrival times, flight identifiers (carrier, flight number, tail number), origin and destination airport information, cancellation and diversion indicators, and delay breakdowns by cause (carrier, weather, NAS, security, late aircraft).

2. **Weather Data**: Hourly meteorological observations from NOAA weather stations including temperature measurements (dry bulb, dew point, wet bulb), precipitation amounts, wind speed and direction, visibility, atmospheric pressure (sea level, station, altimeter settings), relative humidity, and sky condition descriptions. Weather data is joined to flights based on the nearest weather station to each origin airport and the timestamp closest to scheduled departure.

3. **Stations Data**: Metadata about weather station locations including latitude/longitude coordinates, elevation, station identifiers (ICAO codes), and distance calculations between weather stations and airports. This enables proper spatial joining of weather observations to flight records.

**Target Variable:**

Our primary target variable is **DEP_DEL15**, a binary indicator where 1 represents a departure delay of 15 minutes or more, and 0 represents an on-time departure (less than 15 minutes delay). This 15-minute threshold is the industry-standard definition used by the Department of Transportation for reporting airline performance. In our 3-month sample, approximately **20% of flights are delayed** (DEP_DEL15 = 1), while **80% depart on-time** (DEP_DEL15 = 0), revealing significant class imbalance that we must address in our modeling approach.

#### 3.3.2 Data Ingestion and Checkpoint Strategy:

**Ingestion Approach:**

We read data directly into Apache Spark using `spark.read.csv()` or `spark.read.parquet()` depending on the source format. This approach leverages Spark's lazy evaluation, allowing us to work with large datasets efficiently without loading everything into memory at once. For our initial development, we use the 3-month CSV sample; for production scaling, we will convert to Parquet format for improved performance and compression.

**Checkpoint Strategy:**

To maintain an efficient and collaborative workflow, we implement a comprehensive checkpointing strategy using our group folder: `dbfs:/mnt/mids-w261/student-groups/Group_04_04/`

Our checkpoint structure includes:

1. **Raw Data Checkpoint** (`otpw_3m.parquet`): Initial conversion of CSV to Parquet format with inferred schema
2. **Cleaned Data Checkpoint** (`otpw_3m_clean.parquet`): After removing high-missingness columns, duplicates, cancelled/diverted flights, and data leakage features
3. **Imputed Data Checkpoint** (`otpw_3m_imputed.parquet`): After tailored imputation strategies for missing values
4. **Feature Engineering Checkpoint** (`otpw_3m_featured.parquet`): After creating temporal, lag-based, rolling, and weather features
5. **Modeling-Ready Checkpoint** (`otpw_3m_final.parquet`): After encoding, scaling, and final feature selection

This checkpointing strategy provides several benefits:
- **Resumability**: Team members can start from any intermediate stage without re-running earlier steps
- **Debugging**: Errors can be traced to specific pipeline stages
- **Collaboration**: Multiple team members can work on different stages simultaneously
- **Version Control**: Each checkpoint represents a stable state of data transformation
- **Computational Efficiency**: Expensive operations (joins, aggregations) are computed once and cached

**Scalability Considerations:**

As we scale from the 3-month sample (1.4M flights) to larger datasets:
- 6-month sample: ~2.8M flights
- 1-year sample: ~5.6M flights  
- Full 5-year dataset: ~28M flights

We will monitor Spark job performance, partition data appropriately, and adjust cluster resources as needed to maintain reasonable processing times.

#### 3.3.4 Airport–Weather Integration Plan (Data Issues, As-Of Logic & Rationale)

Our exploratory review of the four source tables (flights, airport codes, weather, and station metadata) showed that the **biggest blockers are not in the flights themselves but in the lookup tables we need to join to**. The flights table already carries clean IATA airport codes (`ORIGIN`, `DEST`) and consistent date fields (`FL_DATE`, `YEAR`, `MONTH`, …), so it is a good factual backbone. However, our main airport codes file does **not** include time zones and sometimes only stores geolocation as a single `coordinates` string (`"lon, lat"`). At the same time, the external GitHub airport list **does** provide `timezone`, and often better lat/lon, but it doesn’t perfectly align 1:1 with our current codes file. This makes a direct “flights → airports → weather” join fragile, because we would be mixing two partially-overlapping airport catalogs. To fix this, we will first build **one master airport dimension** by joining the GitHub timezones and geolocation into the codes we already use in flights, and we will coalesce coordinates so that every IATA that appears as an origin/destination ends up with: **(a)** a timezone, **(b)** a lat/lon pair, and **(c)** a human-readable name.

The second major issue is at the **weather** side: NOAA data is hourly and station-based (`STATION`, `DATE`, 100+ weather features), not airport-based. That means there is no native key to connect “ATL, SFO, ORD…” directly to a weather row. Also, stations and airports don’t share the exact same IDs: stations use a different identifier (and sometimes we need to normalize with the `stations.csv` file). On top of that, **weather is in UTC** while our flights are in **local airport time**, so if we don’t add airport timezones first, we can’t reliably pick “the weather hour that corresponds to this departure.” When we inspected the joined sample, we also saw that some rows failed **only because the join was done at `HH:MM` (e.g. `01:59`) while NOAA reports at `HH:00`** — the station was correct, the time bucket was too strict. To solve this, we will: (1) compute airport → nearest-(1..3)-station pairs using the unified lat/lon we just created, (2) store that in a small bridge table (`airport_weather_station`), (3) when we enrich flights, convert flight times to UTC using the airport’s timezone, and **(4) floor/lag to the latest available weather hour** before joining. This “as-of” join pattern gives us a repeatable flow we can scale from the 3-month sample up to 2015–2021.

**Key data problems we identified:**
- Our original airport codes file **lacks time zones**, so flight local times cannot be aligned to UTC weather.
- Airport geolocation is sometimes packed as a single text field (`coordinates`), so we must **parse and standardize lat/lon**.
- Weather rows are **station-based, not airport-based**, so we must create an extra **airport → station** bridge.
- Stations may come from **two slightly different sources** (`weather.csv` vs `stations.csv`), so we need **ID normalization**.
- Not all flights’ origin/dest codes are guaranteed to appear in the GitHub airport list, so we will **fallback to the original codes file** to avoid losing rows.
- Some flights depart at **odd minutes** (e.g. `01:55`, `01:59`) where no weather row exists; we must **date-trunc to hour and/or fallback 1 hour back** to avoid nulls.
- Destination weather can **leak future information**; we will only keep destination observations where `obs_time_utc ≤ prediction_utc`.

## 5. Machine Learning Algorithms (planned, scalable, leakage-aware):
Our task is to decide, **two hours before scheduled departure**, whether a flight will leave **15 minutes late or more** (`DEP_DEL15 = 1`). In the 3-month OTPW sample this is clearly an **imbalanced** classification problem: roughly 1 in 5 flights is delayed, 4 in 5 are on time. That profile is the same one reported in most BTS/NOAA-based flight-delay papers, and in those studies **tree and boosted models** usually outperform simple linear models once you add schedule, airport, and weather features. Because we ultimately have to repeat this on the **1-year** and then **multi-year** OTPW (tens of millions of rows), we are deliberately choosing algorithms that (i) work well on **structured/tabular** data such as “time of day, carrier, origin/dest, weather-as-of,” (ii) already exist in **Spark/MLlib** or have known distributed versions, and (iii) can be **retrained** once we replace the prejoined OTPW with **our own flights → airport → station → weather** joins in Phase 2. In other words, we want something that is good enough now, but won’t block us later when the data shape improves.

#### 5.1 Baseline for Phase 1

For Phase 1 we only need to show that **(a)** we have a leakage-free feature set that respects the **T–2h** rule, and **(b)** a sensible classifier actually learns signal from it. To do that we will start with **Logistic Regression in Spark**. It is fast, easy to explain to an airline stakeholder, and it gives us clean probabilities that we can later threshold. 

The logistic regression model predicts the probability of delay as:

$$
P(y=1 \mid \mathbf{x}) = \frac{1}{1 + e^{-(\beta_0 + \boldsymbol{\beta}^T \mathbf{x})}}
$$

where **x** represents our feature vector and **β** are the learned coefficients.

On the data side, this model will only see features that are truly available **at or before two hours prior to departure** (scheduled times, carrier, origin/destination, calendar features such as day of week or month, plus the "as-of" weather features we already described). Because the target is imbalanced we will turn on **class weights**, which modifies the loss function to:

$$
L = -\frac{1}{n} \sum_{i=1}^{n} w_i \left[ y_i \log(\hat{y}_i) + (1-y_i) \log(1-\hat{y}_i) \right]
$$

where the weights are assigned as:

$$
w_i = \begin{cases} 
w_{\text{delayed}} & \text{if } y_i = 1 \\
w_{\text{ontime}} & \text{if } y_i = 0
\end{cases}
$$

Alternatively, we will use **threshold tuning** so that the model does not collapse to "always on time." The **primary metric** we will report is **F₂**, not F₁, because in operations **missing a real delay is worse than raising a false alert**; F₁ and PR-AUC will still be reported so we can compare to the published work.

Right after that we will keep a **very dumb regression baseline** just for orientation: for those flights that are actually delayed (or that the classifier tags as delayed) we will fit a plain **Linear Regression** on delay minutes:

$$
\text{Delay}_{\text{minutes}} = \alpha_0 + \boldsymbol{\alpha}^T \mathbf{x} + \epsilon
$$

and we will compare it with an even simpler **"predict the average delay"** constant model:

$$
\hat{y} = \bar{y} = \frac{1}{n_{\text{delayed}}} \sum_{i: y_i = 1} \text{Delay}_{\text{minutes}, i}
$$

This restores the original intent in our plan ("linear reg or average delay") and gives Phase 2 something to beat when we bring in boosted regressors.

Finally, still inside Phase 1, we will run a **small Random Forest** (or shallow GBT) on the same leakage-free table. For Random Forest classification, we use **Gini impurity** as the splitting criterion:

$$
\text{Gini} = 1 - \sum_{i=1}^{C} p_i^2
$$

where **C** is the number of classes and **p_i** is the proportion of class **i** at a given node. For Gradient Boosting classification, we minimize **cross-entropy loss**:

$$
L = -\sum_{i=1}^{n} [y_i \log(\hat{y}_i) + (1-y_i) \log(1-\hat{y}_i)]
$$

The only purpose here is to show that the 3-month OTPW sample really contains **nonlinear interactions** — for example, that delays rise faster in the late afternoon **at** congested hubs **under** low visibility. RF/GBT on Spark have already been used at this scale in airline use cases, so we're staying inside the envelope.

#### 5.2 Preferred approach: two-stage pipeline

For the actual project we prefer a **two-stage** design. This follows the pattern used in several recent flight-delay and transport-delay papers: first decide **"will it be delayed?"**, and only then estimate **"by how much?"** The reason is simple: only about **20%** of the flights are interesting from the delay-minutes point of view. If we train a regressor on everyone, 80% of the rows are "boring" and we waste compute on the cluster.

Formally, our two-stage approach can be expressed as:

$$
\text{Stage 1: } \hat{y}_{\text{binary}} = f_{\text{classifier}}(\mathbf{x})
$$

$$
\text{Stage 2: } \hat{y}_{\text{minutes}} = \begin{cases} 
g_{\text{regressor}}(\mathbf{x}) & \text{if } \hat{y}_{\text{binary}} = 1 \\
0 & \text{if } \hat{y}_{\text{binary}} = 0
\end{cases}
$$

In **Stage 1** we will keep the **binary classification** task. We will start with Logistic Regression to have a transparent baseline, and then add a **tree-ensemble model** (Random Forest / GBT / XGBoost-style, depending on what we run first in Databricks). All of these will be trained with **time-series-aware cross-validation** — "train on earlier months, validate on later months" — to make sure we never peek at future weather or future airport congestion. The **primary metric** here remains **F₂** (recall is 4× more important than precision at this stage):

$$
F_2 = 5 \cdot \frac{\text{Precision} \cdot \text{Recall}}{4 \cdot \text{Precision} + \text{Recall}}
$$

and we will also show a **confusion matrix by carrier and by origin airport** so we can see if the model is only good on the big hubs.

In **Stage 2** we will run a **regression model** only on the flights that Stage 1 flags as **delayed**. Here we will move to the usual suspects that do well on tabular data: **gradient-boosted regressor**, **XGBoost regressor**, or **RF regressor**. We will report **MAE** (because it is easy to read: "we're off by ~6 minutes"):

$$
\text{MAE} = \frac{1}{n_{\text{delayed}}} \sum_{i: \hat{y}_{\text{binary},i} = 1} \lvert y_{\text{minutes},i} - \hat{y}_{\text{minutes},i} \rvert
$$

and **RMSE** (because most papers report it):

$$
\text{RMSE} = \sqrt{ \frac{1}{n_{\text{delayed}}} \sum_{i: \hat{y}_{\text{binary},i} = 1} (y_{\text{minutes},i} - \hat{y}_{\text{minutes},i})^2 }
$$

The advantage of doing it this way is that we spend cluster time and feature-engineering effort **only** on the flights that actually need it.

So the story we want in the notebook is: **Phase 1 proves the classification works and is leakage-aware; Phase 2 and 3 add the "how long" part once we scale to 1 year.**

#### 5.3 Alternative: single-regression

We did consider the simpler alternative: **predict delay minutes directly** with a boosted regressor and then **bucket** the output into “on time,” “15–30,” and “30+.” This has been published before and it is attractive because it’s just **one** model. The problem is that **around the 15-minute DOT threshold** these models tend to be a bit less sharp, and you usually end up spending more time on **feature engineering** (trajectory/context, airport-day load, etc.) to recover the performance you can get more easily with a classifier.


#### 5.4 Future-proofing for Phase 3 (time, graph, MLP)

The course eventually wants to see **three** things and we are already aligned with them:

1. **Time series**: we are already doing time-ordered CV and “as-of” weather joins, so we can legitimately say we are treating the data as time-dependent.
2. **Graph**: when we build the airport/carrier network in Phase 2 (nodes = airports or carriers; edges = historical flights; weights = volume or delay rate), we can compute PageRank/degree/centrality and **feed those as extra columns** to the same Stage-1 classifier. That gives us the “graph” angle without rewriting everything.
3. **MLP / NN**: once we have the stable, leakage-free feature table in Spark, we can train **Spark ML’s MLP classifier** on the same target and compare it to the trees.

On top of that — and to keep the spirit of the original plan — we can add an **experimental multi-task MLP** in Phase 3: same shared layers, **two heads** (one for `DEP_DEL15` with binary cross-entropy, one for delay minutes with MSE), and a **weighted sum** of both losses so we can still optimize for “don’t miss delays” while also learning magnitudes. That gives us a modern model to show, but it still respects **all** the Phase-1 leakage rules we just wrote.

## 6. Metrics:
Because our target is **“predict a 15-minute-or-more departure delay two hours before pushback,”** the main risk is **missing a real delay**, not raising an extra alert. The 3-month OTPW sample is imbalanced (≈20% delayed vs. 80% on time), so accuracy alone would hide bad models — a classifier that always says “on time” would score ~0.80 and still be useless for operations. For that reason we will use a **recall-friendly primary metric** and then add a small set of secondary metrics so we can compare across models and across airports.


#### 6.1 Primary classification metric: F₂

We will use **F₂** as the main metric because it weights **recall** higher than precision. First recall the basic definitions:

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

The general F-score is

$$
F_\beta = (1 + \beta^2)\,\frac{\text{Precision} \cdot \text{Recall}}{(\beta^2 \cdot \text{Precision}) + \text{Recall}}
$$

For our case, with \(\beta = 2\),

$$
F_2 = 5 \cdot \frac{\text{Precision} \cdot \text{Recall}}{4 \cdot \text{Precision} + \text{Recall}}
$$

We prefer \(F_2\) because in airline operations a **false negative** (we said “on time” but the flight was actually 15+ minutes late) causes more downstream cost than a **false positive** (we said “will be late” and it wasn’t). We will compute \(F_2\) on **time-ordered validation splits** (train on earlier months → validate on later months) so that the score reflects real-world sequencing.




#### 6.2 Secondary classification metrics: F₁, PR-AUC, ROC-AUC

To stay comparable with the flight-delay literature, we will also report **F₁**, **PR-AUC**, and **ROC-AUC**.

The F₁ score (balanced precision/recall) is

$$
F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

- **F₁**: good for a single headline number when precision and recall matter equally.
- **PR-AUC**: better than ROC-AUC for imbalanced data like ours (20/80).
- **ROC-AUC**: still useful because many published models report it.

This gives us a “compatible” set of metrics even if other papers optimized F₁ instead of \(F_2\).


#### 6.3 Per-segment confusion: carrier and airport

Because we have **carrier** and **origin airport** in the features, we will also break down **confusion matrices** by carrier and by origin for the top-N airports. This is the answer to the stakeholder question: *“does it work on my hub?”* It also helps us detect join problems (e.g. one airport always missing weather, always predicted “on time”).


#### 6.4 Regression metrics (Stage 2 only): MAE first, then RMSE

When we run the **Stage-2 regression** (only on flights that Stage 1 marked as delayed), we will evaluate it with **MAE** and **RMSE**.

Mean Absolute Error:

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \lvert y_i - \hat{y}_i \rvert
$$

Root Mean Squared Error:

$$
\text{RMSE} = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 }
$$

- **MAE** is easier to explain to ops (“we’re off by ~6–8 minutes on average”).
- **RMSE** is good for comparison with prior work and penalizes large errors more.

This fits the two-stage design: **classification** solves the imbalanced detection problem; **regression** is only for the ~20% delayed flights.


#### 6.5 Operational / domain-level views

To keep the “airline consortium” narrative, we will also show two derived, human-readable metrics:

1. **Delay coverage at T–2h**  
   “Out of all flights that actually departed 15+ minutes late, how many did we flag?”  
   This is just **recall** written in airline language.

2. **Alert volume**  
   “How many flights per day would we have pinged?”  
   This is just the **count of positive predictions** and tells ops whether the model is too noisy.

## 8. Pipeline:

In practical terms we will still follow the standard Spark ML pipeline for classification — data cleaning and preprocessing (remove cancelled/diverted flights, drop high-null and post-flight/leakage columns, encode categoricals, scale numerics), optional feature selection (univariate filters or tree-based importances), model training (logistic regression first, then tree/boosting models), model evaluation on a held-out split, and finally saving/checkpointing the model in Databricks. This is the same 5-step pattern described in the Spark MLlib docs and it maps cleanly to our project; the only differences in our case are that (i) all features must be known at or before **T–2h**, (ii) the train/validation split is **time-ordered** rather than random, and (iii) we add a **second stage** for delay-minutes regression but only for the flights that Stage 1 predicts as delayed.

## 10. Open Issues or Problems:


10) Open Issues or Problems

1. Airport–weather join is not fully proven at scale  
   In Phase 1 we used the prejoined OTPW table from Databricks, so the hard part (airport → station → hourly weather → as-of logic) was hidden. For Phase 2 we want to “own” that join, but we still need to validate:  
   - that every ORIGIN/DEST in the 12-month OTPW slice has a timezone;  
   - that the unified airport dimension (original codes + GitHub list) covers all active airports;  
   - that the UTC conversion + “floor to latest hour / fallback −1h” logic actually reduces nulls and does not create future leaks;  
   - that we can store and reuse the small airport ↔ nearest-station bridge.  
   Until we run this on the 1-year table, join quality is still an open item.

2. Possible destination-side leakage  
   Our current narrative says “only information available 2 hours before departure,” but OTPW can carry destination weather and destination operational columns that might be timestamped after departure. We need to re-check the feature list to guarantee that for DEST-side data we only keep rows where `obs_time_utc ≤ dep_time_utc` or we simply drop destination weather to stay safe. This is especially important before we scale to multi-year data.

3. High-null / sparse columns policy must be frozen  
   In Phase 1 we said “~100 columns have >90% missingness” and we dropped them, but that decision was made on 3 months of 2015. On 12 months or 5 years, some of those columns may become usable. We need one written rule (e.g. “drop if >85% null across the whole 12-month table, keep if operationally meaningful”) and we need to re-generate the column-classification table from the larger dataset so the notebook and the report stay consistent.

4. Time-ordered validation not yet implemented in all notebooks  
   The text already says “train on earlier months, validate on later months,” but some example code still uses random splits. Before Phase 2 submission we must (i) push a single helper to create time-based splits, (ii) make sure all models in the notebook call that helper, and (iii) re-report F₂ / F₁ / PR-AUC using the time-ordered split so the metrics reflect real-world sequencing.

5. Class-imbalance strategy still needs to be picked  
   We selected F₂ as the primary metric and we motivate it well, but we haven’t locked the actual imbalance technique: class weights in Logistic Regression, undersampling the majority, or threshold tuning on the validation set. We should benchmark these three on the 3-month table first, and only then port the chosen strategy to the 12-month run.

6. Scaling and checkpoint storage on Databricks  
   The checkpoint strategy is designed (raw → cleaned → imputed → featured → modeling-ready), but we have not yet tested disk usage / cluster runtime for the 12-month or multi-year OTPW tables. We need to confirm that the group folder can hold all intermediate parquet files, and that expensive joins (flight + airport + station + weather) are checkpointed before modeling so every team member does not have to re-run them.

7. Graph / network features not materialized yet  
   The report already promises “airport/carrier network in Phase 2” and using PageRank/degree/centrality as extra features, but we have not yet: (i) built the edges table from historical flights, (ii) chosen node granularity (airport vs carrier vs airport-carrier), or (iii) shown that these features actually improve F₂. This is an open item for Carlos + Anabel in Phase 2–3.

8. Stage-2 regression depends on Stage-1 output  
   We said we will only run MAE/RMSE on flights predicted as delayed, but we have not yet implemented the actual two-step notebook flow (predict → filter delayed → fit regressor). We need to verify that the filtered sample remains large enough across months and airports; otherwise MAE will be unstable and not comparable to the published papers we cite.

9. Reproducibility / HTML export / notebook order  
   Because the team will rotate leaders every week, we need to lock the section order (Data → EDA → ML Strategy → Metrics → Timeline → Pipeline) and add a short “run order” cell at the top so that whoever is the weekly owner can re-export to HTML without breaking references to figures, Gantt, or the pipeline diagram.
