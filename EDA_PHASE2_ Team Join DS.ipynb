```python
from pyspark.sql import functions as F

total = df.count()

target_summary = (
    df.withColumn(
        "DEP_DEL15_status",
        F.when(F.col("DEP_DEL15").isNull(), "Null")
         .when(F.col("DEP_DEL15") == 1, "Delayed (15 min)")
         .otherwise("On Time (<15 min)")
    )
    .groupBy("DEP_DEL15_status")
    .agg(F.count("*").alias("count"))
    .withColumn("percentage", F.round(F.col("count")/F.lit(total)*100, 2))
    .orderBy("DEP_DEL15_status")
)

display(target_summary)
```

```python
# Remove rows where target variable is null
df = df.filter(df.DEP_DEL15.isNotNull())

print(f"After removing null targets: {df.count():,} rows remain.")
```

```python
total = df.count()

# Count non-null values for every column
non_null_agg = df.agg(*[F.count(c).alias(c) for c in df.columns])

# Reshape into (column, non_null)
nulls_long = non_null_agg.selectExpr(
    "stack({}, {}) as (column, non_null)".format(
        len(df.columns),
        ", ".join([f"'{c}', `{c}`" for c in df.columns])
    )
).withColumn("null_count", F.lit(total) - F.col("non_null")) \
 .withColumn("null_pct", F.round(F.col("null_count") / F.lit(total) * 100, 2)) \
 .orderBy(F.desc("null_pct"))

display(nulls_long)
```

```python
# Drop columns with excessive missingness (>80%), always keep the target
threshold = 80.0
cols_to_drop = [r["column"] for r in nulls_long.filter(F.col("null_pct") > threshold).collect()
                if r["column"] != "DEP_DEL15"]

print(f"Drop {len(cols_to_drop)} columns (> {threshold}% null).")
df = df.drop(*cols_to_drop)
print(f"Remaining columns: {len(df.columns)}")
```

```python
# ============================================================
#   CLEAN SAMPLE + ROUTE STATS + EXEC MAP B + EXEC MAP C
# ============================================================

import pyspark.sql.functions as F
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from plotly.colors import sample_colorscale

# ------------------------------------------------------------
# 1) USE df_eda IF IT EXISTS, ELSE USE df
# ------------------------------------------------------------
df_map_src = df_eda if "df_eda" in globals() else df

# ------------------------------------------------------------
# 2) CREATE LIGHTWEIGHT SAMPLE FOR EDA VISUALS
# ------------------------------------------------------------
df_map_base = (
    df_map_src.select(
        "ORIGIN","DEST","DEP_DEL15",
        "origin_airport_lat","origin_airport_lon",
        "dest_airport_lat","dest_airport_lon"
    )
    .filter(
        F.col("origin_airport_lat").isNotNull() &
        F.col("origin_airport_lon").isNotNull() &
        F.col("dest_airport_lat").isNotNull() &
        F.col("dest_airport_lon").isNotNull()
    )
    .sample(0.15, seed=42)   # 15% EDA sample
    .repartition(200)
    .cache()
)

df_map_base.count()  # materialize


# ------------------------------------------------------------
# 3) BUILD ROUTE STATS ON EDA SAMPLE
# ------------------------------------------------------------
routes = (
    df_map_base
    .groupBy(
        "ORIGIN","DEST",
        "origin_airport_lat","origin_airport_lon",
        "dest_airport_lat","dest_airport_lon"
    )
    .agg(
        F.count("*").alias("num_flights"),
        (F.mean("DEP_DEL15") * 100).alias("delay_rate_pct")
    )
)

avg_delay = routes.agg(F.mean("delay_rate_pct")).first()[0]


# ------------------------------------------------------------
# 4) MAP B  Worst High-Volume Routes (Impact = Volume  Delay)
# ------------------------------------------------------------
routes_b = routes.withColumn(
    "impact_score", F.col("num_flights") * F.col("delay_rate_pct")
)
pdf_b = routes_b.orderBy(F.desc("impact_score")).limit(12).toPandas()

# Color by delay (Viridis -> hex)
dvals = pdf_b["delay_rate_pct"].astype(float)
dmin, dmax = dvals.min(), dvals.max()
d_norm = (dvals - dmin) / (dmax - dmin + 1e-9)
pdf_b["color"] = [sample_colorscale("Viridis", float(x))[0] for x in d_norm]

# Width by volume
min_w, max_w = 2, 10
vol = pdf_b["num_flights"].astype(float)
v_norm = (vol - vol.min()) / (vol.max() - vol.min() + 1e-9)
pdf_b["width"] = min_w + v_norm * (max_w - min_w)

# Airports (label hubs only)
airports_b = pd.concat([
    pdf_b[["ORIGIN","origin_airport_lat","origin_airport_lon"]]
        .rename(columns={"ORIGIN":"AIRPORT","origin_airport_lat":"lat","origin_airport_lon":"lon"}),
    pdf_b[["DEST","dest_airport_lat","dest_airport_lon"]]
        .rename(columns={"DEST":"AIRPORT","dest_airport_lat":"lat","dest_airport_lon":"lon"})
]).drop_duplicates("AIRPORT")

top_airports = (
    pd.concat([pdf_b["ORIGIN"], pdf_b["DEST"]])
      .value_counts()
      .head(8)
      .index
)
airports_b["label"] = airports_b["AIRPORT"].where(airports_b["AIRPORT"].isin(top_airports), "")

# Clean NE label nudges
nudges = {
    "JFK": (0.25, -0.15),
    "LGA": (0.15,  0.20),
    "EWR": (-0.25, 0.10),
    "BOS": (0.20,  0.25),
    "DCA": (-0.15,-0.20),
    "LAX": (0.10,-0.10),
    "SFO": (-0.10,0.10)
}
airports_b["lat_adj"] = airports_b.apply(lambda r: r["lat"] + nudges.get(r["AIRPORT"], (0,0))[1], axis=1)
airports_b["lon_adj"] = airports_b.apply(lambda r: r["lon"] + nudges.get(r["AIRPORT"], (0,0))[0], axis=1)

# Build Map B
figB = go.Figure()

for _, r in pdf_b.iterrows():
    figB.add_trace(go.Scattergeo(
        lon=[r["origin_airport_lon"], r["dest_airport_lon"]],
        lat=[r["origin_airport_lat"], r["dest_airport_lat"]],
        mode="lines",
        line=dict(width=float(r["width"]), color=r["color"]),
        opacity=0.7,
        hoverinfo="text",
        text=(
            f'{r["ORIGIN"]}{r["DEST"]}<br>'
            f'Flights: {int(r["num_flights"]):,}<br>'
            f'Delay rate: {r["delay_rate_pct"]:.1f}%<br>'
            f'Impact score: {r["impact_score"]:.0f}'
        ),
        showlegend=False
    ))

figB.add_trace(go.Scattergeo(
    lon=airports_b["lon_adj"], lat=airports_b["lat_adj"],
    mode="markers+text",
    text=airports_b["label"],
    textposition="top center",
    marker=dict(size=7),
    showlegend=False
))

figB.update_layout(
    title="Map B  Worst High-Volume Delay Corridors (Impact = Volume  Delay)",
    geo=dict(scope="usa", projection_type="albers usa",
             showland=True, landcolor="rgb(245,245,245)"),
    height=650
)

display(figB)


# ------------------------------------------------------------
# 5) MAP C  Busiest Routes with Delay Overlay (Exec Clean)
# ------------------------------------------------------------
pdf_c = routes.orderBy(F.desc("num_flights")).limit(12).toPandas()

# Color by delay
dvals_c = pdf_c["delay_rate_pct"].astype(float)
dmin_c, dmax_c = dvals_c.min(), dvals_c.max()
d_norm_c = (dvals_c - dmin_c) / (dmax_c - dmin_c + 1e-9)
pdf_c["color"] = [sample_colorscale("Viridis", float(x))[0] for x in d_norm_c]

# Width by volume
vol_c = pdf_c["num_flights"].astype(float)
v_norm_c = (vol_c - vol_c.min()) / (vol_c.max() - vol_c.min() + 1e-9)
pdf_c["width"] = min_w + v_norm_c * (max_w - min_w)

# Airport labels for Map C
airports_c = pd.concat([
    pdf_c[["ORIGIN","origin_airport_lat","origin_airport_lon"]]
        .rename(columns={"ORIGIN":"AIRPORT","origin_airport_lat":"lat","origin_airport_lon":"lon"}),
    pdf_c[["DEST","dest_airport_lat","dest_airport_lon"]]
        .rename(columns={"DEST":"AIRPORT","dest_airport_lat":"lat","dest_airport_lon":"lon"})
]).drop_duplicates("AIRPORT")

top_airports_c = (
    pd.concat([pdf_c["ORIGIN"], pdf_c["DEST"]])
      .value_counts()
      .head(8)
      .index
)
airports_c["label"] = airports_c["AIRPORT"].where(airports_c["AIRPORT"].isin(top_airports_c), "")

airports_c["lat_adj"] = airports_c.apply(lambda r: r["lat"] + nudges.get(r["AIRPORT"], (0,0))[1], axis=1)
airports_c["lon_adj"] = airports_c.apply(lambda r: r["lon"] + nudges.get(r["AIRPORT"], (0,0))[0], axis=1)

# Build Map C
figC = go.Figure()

for _, r in pdf_c.iterrows():
    figC.add_trace(go.Scattergeo(
        lon=[r["origin_airport_lon"], r["dest_airport_lon"]],
        lat=[r["origin_airport_lat"], r["dest_airport_lat"]],
        mode="lines",
        line=dict(width=float(r["width"]), color=r["color"]),
        opacity=0.7,
        hoverinfo="text",
        text=(
            f'{r["ORIGIN"]}{r["DEST"]}<br>'
            f'Flights: {int(r["num_flights"]):,}<br>'
            f'Delay rate: {r["delay_rate_pct"]:.1f}%'
        ),
        showlegend=False
    ))

figC.add_trace(go.Scattergeo(
    lon=airports_c["lon_adj"], lat=airports_c["lat_adj"],
    mode="markers+text",
    text=airports_c["label"],
    textposition="top center",
    marker=dict(size=7),
    showlegend=False
))

figC.update_layout(
    title="Map C  Busiest Routes with Delay Severity Overlay",
    geo=dict(scope="usa", projection_type="albers usa",
             showland=True, landcolor="rgb(245,245,245)"),
    height=650
)

display(figC)
```

```python
import pyspark.sql.functions as F
import plotly.express as px
import pandas as pd
import numpy as np

# --- Ensure departure hour exists ---
hour_col = "departure_hour" if "departure_hour" in df.columns else None
if hour_col is None:
    df_tmp = df.withColumn(
        "departure_hour",
        F.floor(F.col("CRS_DEP_TIME") / 100).cast("int")
    )
    hour_col = "departure_hour"
else:
    df_tmp = df

# --- Time blocks (ASCII-only labels) ---
df_tmp = df_tmp.withColumn(
    "time_block",
    F.when((F.col(hour_col) >= 0) & (F.col(hour_col) < 6),  "Early AM (12-6a)")
     .when((F.col(hour_col) >= 6) & (F.col(hour_col) < 12), "Morning (6a-12p)")
     .when((F.col(hour_col) >= 12) & (F.col(hour_col) < 17), "Afternoon (12-5p)")
     .when((F.col(hour_col) >= 17) & (F.col(hour_col) < 21), "Evening (5-9p)")
     .otherwise("Night (9p-12a)")
)

# --- Aggregate delay rate ---
heat = (
    df_tmp
    .filter(F.col("time_block").isNotNull() & F.col("DAY_OF_WEEK").isNotNull())
    .groupBy("time_block", "DAY_OF_WEEK")
    .agg(
        F.count("*").alias("num_flights"),
        (F.mean("DEP_DEL15") * 100).alias("delay_rate_pct")
    )
    .filter(F.col("num_flights") >= 500)
)

pdf_heat = heat.toPandas()

block_order = [
    "Early AM (12-6a)",
    "Morning (6a-12p)",
    "Afternoon (12-5p)",
    "Evening (5-9p)",
    "Night (9p-12a)"
]
day_order = [1, 2, 3, 4, 5, 6, 7]
day_labels = {1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat", 7: "Sun"}

pivot = (
    pdf_heat
    .pivot(index="time_block", columns="DAY_OF_WEEK", values="delay_rate_pct")
    .reindex(index=block_order, columns=day_order)
)

pivot.columns = [day_labels[d] for d in pivot.columns]
pivot = pivot.astype(float)

# --- Find top 3 hotspots ---
z = pivot.values
flat_idx = np.argsort(np.nan_to_num(z, nan=-1).ravel())[::-1][:3]
hotspots = [np.unravel_index(i, z.shape) for i in flat_idx]

# --- Plot with executive-friendly colormap ---
fig = px.imshow(
    pivot,
    color_continuous_scale="Viridis",
    labels=dict(x="Day of week", y="Time of day", color="Delay rate (%)"),
    aspect="auto",
    title="Delay Hotspots (Predictable Windows, Not Random)"
)

# --- Add contrast-aware percent labels ---
threshold = np.nanmean(z)

for i, row in enumerate(pivot.index):
    for j, col in enumerate(pivot.columns):
        val = pivot.iloc[i, j]
        if not np.isnan(val):
            text_color = "white" if val < threshold else "black"
            fig.add_annotation(
                x=col,
                y=row,
                text=f"{val:.1f}%",
                showarrow=False,
                font=dict(size=14, color=text_color)
            )

# --- Add subtle hotspot outlines ---
for (i, j) in hotspots:
    fig.add_shape(
        type="rect",
        x0=j - 0.5, x1=j + 0.5,
        y0=i - 0.5, y1=i + 0.5,
        line=dict(width=2, color="white")
    )

fig.update_layout(
    height=520,
    xaxis_side="top",
    title_font_size=22
)

fig.show()
```

Additionally, the **no-precipitation group (0.00 inches)** was removed since it represents over 90% of all flights and dominates the visualization.  

```python
from pyspark.sql.types import DecimalType

BIN = 0.01

df_precip_num = (
    df_clean
      .withColumn("precip_raw", F.col("HourlyPrecipitation").cast("string"))
      .withColumn(
          "precip_num",
          F.when(F.col("precip_raw") == "T", 0.0)
           .otherwise(F.regexp_replace("precip_raw", "[^0-9\\.]", ""))
           .cast("double")
      )
)
precip_binned = (
    df_precip_num
      .filter(F.col("precip_num").isNotNull())             
      .withColumn("bin_idx", F.floor(F.col("precip_num")/F.lit(BIN)).cast("int"))
      .withColumn("Precip_bin",
                  (F.col("bin_idx")*F.lit(BIN)).cast(DecimalType(4,2))) 
      .groupBy("bin_idx","Precip_bin")
      .agg(
          F.count("*").alias("num_flights"),
          F.round(F.mean("DEP_DEL15")*100, 2).alias("delay_rate_pct")
      )
      .orderBy("bin_idx")
      .drop("bin_idx")
)

display(precip_binned)
```

```python
# Filter out bins with very few samples and precipitation = 0
precip_filtered = precip_binned.filter(
    (F.col("Precip_bin") > 0) & (F.col("num_flights") >= 100)
)
display(precip_filtered)
```

### 2.5 Distance Impact on Delays
Understanding how flight distance and route characteristics influence delay rates helps identify whether delays are primarily due to operational congestion or weather dependencies.


#### 2.5.1 Distance Delay Patterns

```python
# Calculate delay rate by distance bin (every 250 miles)

distance_delay = (
    df_clean.filter(F.col("DISTANCE").isNotNull())
      .withColumn("distance_bin", (F.floor(F.col("DISTANCE") / 250) * 250))
      .groupBy("distance_bin")
      .agg(
          F.count("*").alias("num_flights"),
          F.round(F.mean("DEP_DEL15") * 100, 2).alias("delay_rate_pct")
      )
      .orderBy("distance_bin")
)

display(distance_delay)
```

```python
import pyspark.sql.functions as F
import plotly.express as px
import plotly.graph_objects as go

# Use your df name here
df = df_clean

# Add scheduled departure hour (CRS_DEP_TIME like 0..2359)
df = df.withColumn(
    "CRS_DEP_HOUR",
    F.floor(F.col("CRS_DEP_TIME")/100).cast("int")
)

# A reusable delayed flag
df = df.withColumn(
    "is_delayed",
    F.when(F.col("DEP_DEL15") >= 1, 1).otherwise(0)
)
```
