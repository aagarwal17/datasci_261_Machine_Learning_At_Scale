```python
# This cell defines a function to copy a picture file from a user's personal workspace folder to a shared team folder in DBFS.
# 
# Functionality:
# - The function `copy_picture_to_team_folder` takes a picture file name and a user's email address.
# - It constructs the source path in the user's workspace and the destination path in the team folder.
# - It uses `dbutils.fs.cp` to copy the file from the user's folder to the team folder.
# - The example usage copies "ML_pipeline_phase_2.png" from "carlos.schrupp@berkeley.edu"'s workspace to the team folder.
#
# Instructions:
# 1. Ensure the picture file exists in your workspace under `/Workspace/Users/<your_email>/pictures/`.
# 2. Call `copy_picture_to_team_folder(picture_name, user_email)` with the correct file name and your email address.
#    Example:
#    picture = "your_picture.png"
#    email = "your.email@domain.com"
#    copy_picture_to_team_folder(picture, email)
# 3. The file will be copied to `dbfs:/student-groups/Group_4_4/` for team access.
```

```python
def copy_picture_to_team_folder(picture_name, user_email):
    source_path = f"file:/Workspace/Users/{user_email}/pictures/{picture_name}"
    dest_path = f"dbfs:/student-groups/Group_4_4/Charts/{picture_name}"
    dbutils.fs.cp(source_path, dest_path)

email = "carlos.schrupp@berkeley.edu"
picture = "classification.png"
copy_picture_to_team_folder(picture, email)
```
